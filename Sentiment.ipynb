```python
nltk.download('punkt')
url = "https://www.gutenberg.org/files/768/768.txt"
response = request.urlopen(url)
raw = response.read().decode('utf8')
tokens = nltk.word_tokenize(raw)
```

```python
words = [w.lower() for w in tokens]
```

```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
tokens = [w for w in words if w not in stopwords]
```

```python
tokens = [t for t in tokens if t.isalpha()]
```

```python
wordList = defaultdict(list)
```

```python
emotionList = defaultdict(list)
```

```python
filepath = 'NRC-Emotion-Lexicon-v0.92/NRC-Emotion-Lexicon-Wordlevel-v0.92.txt'
```

```python
with open(filepath, 'r') as f:
    reader = csv.reader(f, delimiter='\t')
    headerRows = [i for i in range(0, 46)]
    for row in headerRows:
        next(reader)
    for word, emotion, present in reader:
        if int(present) == 1:
            wordList[word].append(emotion)
            emotionList[emotion].append(word)
```

```python
def generate_emotion_count(tokens):
    emoCount = Counter()
    for t in tokens:
        if len(wordList[t]) > 0:
            emoCount += Counter(wordList[t])
    return emoCount
            
```

```python
emotionCounts = generate_emotion_count(tokens)
```
